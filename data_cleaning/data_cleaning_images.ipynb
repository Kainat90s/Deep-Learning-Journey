1.  Corrupt Images Remove karna

Definition:
Corrupt image woh hoti hai jo open hi nahi hoti — file andar se toot gayi hoti hai, size zero hoti hai, ya properly save nahi hui hoti!

Real life example:
Socho tumne 1000 cat/dog images download ki —
Kuch images download hote hote internet cut ho gaya —
Woh images corrupt ho gayi! ❌
Agar yeh corrupt images model ko dein:

Model confuse hoga ❌
Error aayega ❌

```python
from PIL import Image
import os

# Folder jahan images hain
image_folder = "cat"

# Folder banao test ke liye
os.makedirs(image_folder, exist_ok=True)

# Corrupt images dhundho
corrupt = []

for img_name in os.listdir(image_folder):
    img_path = os.path.join(image_folder, img_name)
    try:
        img = Image.open(img_path)
        img.verify()  # image theek hai?
      
    except:
        corrupt.append(img_name)
        os.remove(img_path)  # corrupt image delete karo
        print(f"Corrupt image remove ki: {img_name}")

print(f"Total corrupt images: {len(corrupt)}")
```

Duplicate Images Remove karna

Definition:
Duplicate image woh hoti hai jo dataset mein 2 baar ya zyada baar mojood hoti hai — bilkul same image!

Real life example:
Socho tumne 1000 cat images download ki —
100 images accidentally 2 baar download ho gayi ❌
Agar yeh duplicates model ko dein:

Model usi image ko zyada baar dekhe ga
Bias aayega — model usi cheez ko zyada seekhe ga ❌

```python
import hashlib
import os

image_folder = r"C:\Users\PMLS\OneDrive\Desktop\images"
hashes = {}
duplicates = []

for img_name in os.listdir(image_folder):
    img_path = os.path.join(image_folder, img_name)
    
    with open(img_path, "rb") as f:
        img_hash = hashlib.md5(f.read()).hexdigest()
        
    
    if img_hash in hashes:
        duplicates.append(img_name)
        os.remove(img_path)
        print(f"Duplicate remove ki: {img_name}")
    else:
        hashes[img_hash] = img_name

print(f"Total duplicates: {len(duplicates)}")
```

 Images resize karna

📌 Images Resize karna kyun zaroori hai?
Definition:
Neural network ko same size ki images chahiye hoti hain — agar alag alag size ki images dein tو model confuse hoga!

```python
from PIL import Image
import os

images = r"C:\Users\PMLS\OneDrive\Desktop\images"
output_folder = r"C:\Users\PMLS\OneDrive\Desktop\images_resized"

# Output folder banao
os.makedirs(output_folder, exist_ok=True)

# Har image resize karo
for img_name in os.listdir(images):
    img_path = os.path.join(images, img_name)
    
    try:
        img = Image.open(img_path)
        img_resized = img.resize((224, 224))
        img_resized.save(os.path.join(output_folder, img_name))
        print(f"Resized: {img_name}")
    except:
        print(f"Skip: {img_name}")

print("Done!")
```

 Pixels Normalize karna
 

Definition:
Image ke pixels ki values 0 to 255 hoti hain — inhe 0 to 1 mein convert karna normalization hai!

Kyun zaroori hai?

Model ke liye chhote numbers handle karna easy hota hai
Training fast hoti hai
Model better seekhta hai ✅

Original pixel = 255
Normalized     = 255 / 255 = 1.0

Original pixel = 128
Normalized     = 128 / 255 = 0.50

Original pixel = 0
Normalized     = 0 / 255 = 0.0

```python
import numpy as np
from PIL import Image

# Image kholo
img = Image.open(r"C:/Users/PMLS/OneDrive/Desktop/images/OIP.webp")

# Numpy array mein badlo
img_array = np.array(img)

print(f"Before normalize: {img_array.min()} to {img_array.max()}")

# Normalize karo — 0 to 1
img_normalized = img_array / 255.0

print(f"After normalize: {img_normalized.min()} to {img_normalized.max()}")
```

255 / 255 = 1.0   ← sabse roشن pixel


0 / 255   = 0.0   ← sabse dark pixel


128 / 255 = 0.50  ← beech wala
