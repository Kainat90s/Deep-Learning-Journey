# Lower casing

Saare letters ko small mein convert karna — taake "Cat" aur "cat" model ke liye same hon!

```python
text="Helow i love you kainat!"
correct=text.lower()
print(correct)
```

```python
text = "Hello! I LOVE Cats and DOGS!"

# Lowercasing
text_lower = text.lower()

print(f"Before: {text}")
print(f"After:  {text_lower}")
```

# Punctuation Removal


Definition:
Text mein se ! ? . , ; : ' " jaise symbols hatana — yeh model ke liye koi meaning nahi rakhte!

Example:
Before: "Hello! I love cats..."
After:  "Hello I love cats"

```python
import re

text = "Hello! I LOVE Cats... and DOGS?"

# Punctuation remove karo
text_clean = re.sub(r'[^\w\s]', '', text) #r raw string. [w mai words and 
#numbers atai hai  s mai spaces inkai ilawa jo b aey delete kro]
#re.sub = find karo aur replace karo
print(f"Before: {text}")
print(f"After:  {text_clean}")
```

```python
import re

text = "kainat is a good girl"

# Pehla match dhundho
result = re.search(r'[^\w]', text)

print(result)
print(result.group())  # jo mila wo dikhao
```

# Stopwords Removal

Definition:
Woh words jo text mein bohat common hain lekin koi meaning nahi rakhte — jaise "is, a, the, an, in"!

Example:
Before: "kainat is a good girl"
After:  "kainat good girl"

Kyun remove karte hain?

Model ke liye unnecessary words hain
Model important words pe focus kare ✅

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

text = "kainat is a very good girl and she loves cats"

# Stopwords ki list
stop_words = set(stopwords.words('english'))

# Words mein todo
words = text.split()

# Stopwords remove karo
clean_words = [word for word in words if word not in stop_words]

clean_text = ' '.join(clean_words)

print(f"Before: {text}")
print(f"After:  {clean_text}")
```

# Tokenization kya hoti hai?

 Tokenization kya hoti hai?
Definition:
Text ko chhote chhote pieces mein todna — jaise words ya sentences mein!

Example:
Before: "kainat is a good girl"
After:  ['kainat', 'is', 'a', 'good', 'girl']

```python
import nltk
nltk.download('punkt', download_dir='C:/nltk_data')
```

```python

from nltk.tokenize import word_tokenize

text = "I don't like cats! She loves them."

# Split
print("Split:",text.split())
w=word_tokenize(text)
# Tokenization
print("Tokenize:",w)
```
