# find Dupllicate images and remove

```python
import os
import hashlib

hash_img={}
dupicate_images=[]
folder=r"C:\Users\PMLS\OneDrive\Desktop\images\cat_folder"
for image_name in os.listdir(folder):
    image_path=os.path.join(folder,image_name)
    try:
        with open(image_path,'rb') as f:
            img=hashlib.md5(f.read()).hexdigest()
           
            if img in hash_img:
                print("dupllicated founded")
                dupicate_images.append(image_name)
                os.remove(image_path)
            else:
                 hash_img[img]=image_name
                
    # except FileNotFoundError:
    #   print("file not found")
    except Exception as e:                       # ✅ koi bhi error pakdo
            print(f"Error: {e}")
print(f"duplicated images : {len(dupicate_images)}")
print(f" unique images: {len(hash_img)}")
```
