Task 4 — Palindrome Check (Medium)

Function check kare word ulta seedha same hai ya nahi:

Input:  "madam"

Output: "madam is Palindrome ✅"


Input:  "hello"

Output: "hello is Not Palindrome ❌"

```python
def palindrom(word):
   
    if word==word[::-1]:
        print("pallindrom")
    else:
        print("not palindrom")
palindrom("madam")
    
```

```python
def palindrome(word):
    if word == word[::-1]:   # ✅ == aur : lagao
        print(f"{word} Palindrome hai ✅")
    else:
        print(f"{word} Palindrome nahi hai ❌")

palindrome("madam")   # Palindrome hai ✅
palindrome("hello")   # Palindrome nahi hai ❌
```
